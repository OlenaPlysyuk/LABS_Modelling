```python
#  1
import matplotlib.pyplot as plt


N = 17

a11 = 0.01 * N
a12 = 0.0001 * N
a21 = 0.0001 * N
a22 = 0.04 * N

x0 = 1000 - 10 * N
y0 = 700 - 10 * N

t0 = 0
T = 150
h = 0.1

steps = int((T - t0) / h) + 1

t_values = [t0]
x_values = [x0]
y_values = [y0]

def f_x(x, y):
    return a11 * x - a12 * x * y

def f_y(x, y):
    return a21 * x * y - a22 * y

for i in range(steps - 1):
    t = t_values[-1]
    x = x_values[-1]
    y = y_values[-1]

    k1_x = h * f_x(x, y) #  
    k1_y = h * f_y(x, y)

    k2_x = h * f_x(x + k1_x / 2, y + k1_y / 2) # 
    k2_y = h * f_y(x + k1_x / 2, y + k1_y / 2)

    k3_x = h * f_x(x + k2_x / 2, y + k2_y / 2) # 
    k3_y = h * f_y(x + k2_x / 2, y + k2_y / 2)

    k4_x = h * f_x(x + k3_x, y + k3_y)
    k4_y = h * f_y(x + k3_x, y + k3_y) # 

    x_next = x + (k1_x + 2 * k2_x + 2 * k3_x + k4_x) / 6
    y_next = y + (k1_y + 2 * k2_y + 2 * k3_y + k4_y) / 6

    t_values.append(t + h)
    x_values.append(x_next)
    y_values.append(y_next)

for i in range(0, len(t_values), int(1/h)):
    print("t =", round(t_values[i], 1), "x =", round(x_values[i], 2), "y =", round(y_values[i], 2))


plt.figure(figsize=(10, 5))
plt.plot(t_values, x_values, label=' x(t)', color='blue')
plt.plot(t_values, y_values, label=' y(t)', color='red')
plt.xlabel(' ()')
plt.ylabel('')
plt.title('   ')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label='  y(x)', color='green')
plt.xlabel(' x')
plt.ylabel(' y')
plt.title('  ')
plt.legend()
plt.grid()
plt.show()

```

```python
# 2

H = 1000 - N
beta = 25 - N
gamma = N

x0 = 900 - N
y0 = 90 - N
z0 = H - x0 - y0

t0 = 0
T = 40
h = 0.1

steps = int((T - t0) / h) + 1

t_values = [t0]
x_values = [x0]
y_values = [y0]
z_values = [z0]

def f_x(x, y):
    return - (beta / H) * x * y

def f_y(x, y):
    return (beta / H) * x * y - (1 / gamma) * y

def f_z(y):
    return (1 / gamma) * y

for i in range(steps - 1):
    t = t_values[-1]
    x = x_values[-1]
    y = y_values[-1]
    z = z_values[-1]

    k1_x = h * f_x(x, y)
    k1_y = h * f_y(x, y)
    k1_z = h * f_z(y)

    k2_x = h * f_x(x + k1_x / 2, y + k1_y / 2)
    k2_y = h * f_y(x + k1_x / 2, y + k1_y / 2)
    k2_z = h * f_z(y + k1_y / 2)

    k3_x = h * f_x(x + k2_x / 2, y + k2_y / 2)
    k3_y = h * f_y(x + k2_x / 2, y + k2_y / 2)
    k3_z = h * f_z(y + k2_y / 2)

    k4_x = h * f_x(x + k3_x, y + k3_y)
    k4_y = h * f_y(x + k3_x, y + k3_y)
    k4_z = h * f_z(y + k3_y)

    x_next = x + (k1_x + 2 * k2_x + 2 * k3_x + k4_x) / 6
    y_next = y + (k1_y + 2 * k2_y + 2 * k3_y + k4_y) / 6
    z_next = z + (k1_z + 2 * k2_z + 2 * k3_z + k4_z) / 6

    t_values.append(t + h)
    x_values.append(x_next)
    y_values.append(y_next)
    z_values.append(z_next)

for i in range(0, len(t_values), int(1/h)):
    print("t =", round(t_values[i], 1), "x =", round(x_values[i], 2), "y =", round(y_values[i], 2), "z =", round(z_values[i], 2))




plt.figure(figsize=(10, 5))
plt.plot(t_values, x_values, label=' x(t)', color='blue')
plt.plot(t_values, y_values, label=' y(t)', color='red')
plt.plot(t_values, z_values, label=' z(t)', color='green')
plt.xlabel(' ()')
plt.ylabel(' ')
plt.title(' ')
plt.legend()
plt.grid()
plt.show()
```
